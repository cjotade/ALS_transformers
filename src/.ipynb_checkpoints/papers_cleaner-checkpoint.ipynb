```python
import os
import re
```

```python
DATA_FOLDER = "./data/papers/"
```

```python
for paper_folder in filter(lambda x: not x.endswith(".zip"), os.listdir(DATA_FOLDER)):
    paper_path = os.path.join(DATA_FOLDER, paper_folder)
    papers = list(filter(lambda x: x.endswith(".tex"), os.listdir(paper_path)))
    for filename in papers:
        filepath = os.path.join(paper_path, filename)
        tex_file = open(filepath, "r").read()
        
        
        #print(re.search(r"\\begin{abstract} (.*?) \\end{abstract}", tex_file))
        #search = re.search(r"\\begin{abstract}.*", tex_file)
        search = re.search(r"(?s)\\begin.*?\\end", tex_file)
        print(tex_file)
        print()
        print(search)
        print()
        print(tex_file[search.start():search.end()])
        
        
    break
```
